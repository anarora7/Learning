```python
def calculate_rectangle_area(length, width):
    """Calculates the area of a rectangle.

    Args:
        length (float or int): The length of the rectangle.
        width (float or int): The width of the rectangle.

    Returns:
        (float or int): The area of the rectangle.

    Raises:
        ValueError: If length or width is negative.
    """
    if length < 0 or width < 0:
        raise ValueError("Length and width cannot be negative.")
    return length * width

print(calculate_rectangle_area(5, 10))
print(calculate_rectangle_area(0, 5))
print(calculate_rectangle_area(3, 4))
```

```python
import inspect

def generate_and_run_pytest_file(main_func, test_funcs_list, filename='test_function.py'):
    """Generates a Python file with a main function and its tests, then runs pytest on it.

    Args:
        main_func (function): The main function to be tested.
        test_funcs_list (list): A list of test functions for the main function.
        filename (str, optional): The name of the file to create. Defaults to 'test_function.py'.
    """
    # Get the source code of the main function
    function_code = inspect.getsource(main_func)

    # Get the source code of the test functions
    test_code_combined = ""
    for test_func in test_funcs_list:
        test_code_combined += inspect.getsource(test_func)

    # Combine the function definition and test functions into the test file content
    test_file_content = f"""
import pytest

{function_code}

{test_code_combined}
"""

    # Write the test code to a temporary file
    with open(filename, 'w') as f:
        f.write(test_file_content)

    # Run pytest on the temporary file
    print(f"Running pytest on {filename}:")
    !pytest {filename}
```

```python
import pytest
from __main__ import calculate_rectangle_area  # corrected import

def test_calculate_rectangle_area_positive_values():
    assert calculate_rectangle_area(5, 10) == 50
    assert calculate_rectangle_area(2.5, 4) == 10.0
    assert calculate_rectangle_area(1, 1) == 1

def test_calculate_rectangle_area_zero_values():
    assert calculate_rectangle_area(0, 10) == 0
    assert calculate_rectangle_area(10, 0) == 0
    assert calculate_rectangle_area(0, 0) == 0

def test_calculate_rectangle_area_negative_length_raises_value_error():
    with pytest.raises(ValueError, match="cannot be negative"):
        calculate_rectangle_area(-1, 5)

def test_calculate_rectangle_area_negative_width_raises_value_error():
    with pytest.raises(ValueError, match="cannot be negative"):
        calculate_rectangle_area(5, -1)

def test_calculate_rectangle_area_both_negative_raises_value_error():
    with pytest.raises(ValueError, match="cannot be negative"):
        calculate_rectangle_area(-3, -7)

def test_calculate_rectangle_area_non_numeric_input_raises_type_error():
    with pytest.raises(TypeError):
        calculate_rectangle_area("abc", 5)
    with pytest.raises(TypeError):
        calculate_rectangle_area(5, "def")
    with pytest.raises(TypeError):
        calculate_rectangle_area("abc", "def")

def test_calculate_rectangle_area_large_values():
    assert calculate_rectangle_area(1000000, 1000000) == 1000000000000
    assert calculate_rectangle_area(999999999, 999999999) == 999999998000000001
```
